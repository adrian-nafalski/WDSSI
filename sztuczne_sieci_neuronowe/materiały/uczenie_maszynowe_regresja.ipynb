# Regression

Regression is a type of supervised learning technique in machine learning that is used to predict a continuous numerical output value based on input data. In regression, the goal is to build a model that can estimate the relationship between a set of input variables, also known as independent variables or features, and a single output variable, also known as the dependent variable or target variable.

Regression models work by fitting a mathematical function to the input data, which can then be used to make predictions on new input data. The mathematical function used to fit the data is typically chosen based on the type of relationship between the input and output variables, and can be linear or non-linear.

Regression models are commonly used in machine learning for a variety of applications, including:

* Predicting future values of a variable: Regression models can be used to make predictions about future values of a variable based on historical data. For example, a regression model can be used to predict future stock prices based on past performance.

* Understanding relationships between variables: Regression models can be used to analyze the relationship between input and output variables and identify factors that may influence the output variable. For example, a regression model can be used to study the relationship between a person's age, income, and education level and their likelihood of purchasing a particular product.

* Feature selection: Regression models can be used to select the most important features or input variables that are most predictive of the output variable. This can help to simplify the model and improve its performance.

In summary, regression is a type of supervised learning technique that is used to predict a continuous numerical output value based on input data. Regression models are widely used in machine learning for a variety of applications, including predicting future values of a variable, understanding relationships between variables, and feature selection.

## Introduction:
In machine learning, regression and logistic regression are two commonly used techniques for predicting numerical and categorical variables, respectively. Regression is used to predict a continuous variable, whereas logistic regression is used to predict a categorical variable.

> **Regression** is a statistical method used to analyze the relationship between a dependent variable and one or more independent variables. It is used to predict the value of a dependent variable based on the values of one or more independent variables. Regression analysis can be used to find the relationship between different variables, identify the key factors affecting the dependent variable, and make predictions about future outcomes.

####Simple Linear Regression:
Simple linear regression is used when there is a linear relationship between the independent variable X and dependent variable Y. The mathematical formula for simple linear regression is:

$Y = \beta_{0} + \beta_{1}X + \epsilon$

where,

* Y is the dependent variable
* X is the independent variable
* $\beta_{0}$ is the intercept
* $\beta_{1}$ is the slope
* $\epsilon$ is the error term.

#### Multiple Linear Regression:
Multiple linear regression is used when there are multiple independent variables X1, X2, X3, ..., Xn, and one dependent variable Y. The mathematical formula for multiple linear regression is:

$Y = \beta_{0} + \beta_{1}X_{1} + \beta_{2}X_{2} + ... + \beta_{n}X_{n} + \epsilon$

where,

* Y is the dependent variable
* $X_1, X_2, X_3, ..., X_n$ are the independent variables
* $\beta_{0}$ is the intercept
* $\beta_{1}$, $\beta_{2}$, $\beta_{3}$, ..., $\beta_{n}$ are the slopes
* $\epsilon$ is the error term.

#### Logistic Regression:

Logistic regression is used when the dependent variable is categorical. It is used to predict the probability of an event occurring. The output of logistic regression is a probability value between 0 and 1. If the probability is greater than or equal to a certain threshold value, the event is predicted to occur; otherwise, it is predicted not to occur.

Binary Logistic Regression:
Binary logistic regression is used when there are only two categories in the dependent variable. The mathematical formula for binary logistic regression is:

$P(Y=1|X) = \frac{1}{1+e^{-(\beta_{0} + \beta_{1}X)}}$

where,

* P(Y=1|X) is the probability of Y=1 given the value of X
* X is the independent variable
* $\beta_{0}$ is the intercept
* $\beta_{1}$ is the slope.

#### Multinomial Logistic Regression:
Multinomial logistic regression is used when there are more than two categories in the dependent variable. The mathematical formula for multinomial logistic regression is:

## Example of linear regression
$y = ax +b$ - here's an example of how linear regression works in Python using the popular Matplotlib and Scikit-learn libraries:

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Generate some sample data
x = np.array([1, 2, 3, 4, 5, 6]).reshape((-1, 1))
y = np.array([2, 4, 5, 6, 7, 8])

# Create a linear regression model and fit the data
model = LinearRegression().fit(x, y)

# Generate predictions on new data
x_new = np.array([7, 8]).reshape((-1, 1))
y_new = model.predict(x_new)

# Plot the original data and the regression line
plt.scatter(x, y)
plt.plot(x, model.predict(x), color='red')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear Regression Example')
plt.show()
```

In this example, we create a simple dataset with six input-output pairs, where the input `x` is a single feature and the output `y` is a continuous numerical value. We then use the Scikit-learn library to create a linear regression model, which we fit to the data using the `fit()` method.

Next, we generate some new data points to test the model on using the predict() method. Finally, we plot the original data as a scatter plot and the regression line as a red line using Matplotlib.

When we run this code, we get a plot that shows the original data points as blue dots and the regression line as a red line.

We can see that the regression line fits the data fairly well, and we can use this model to make predictions on new data points that we haven't seen before.

# Example 2:
A more complex dataset for regression using an existing dataset, the _Boston Housing_ dataset, and the popular Matplotlib and Scikit-learn libraries:

```python
from sklearn.datasets import fetch_california_housing
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Load the California Housing dataset
X, y = fetch_california_housing(return_X_y=True)

# Create a linear regression model and fit the data
model = LinearRegression().fit(X, y)

# Generate predictions on new data
y_pred = model.predict(X)

# Plot the actual and predicted values
plt.scatter(y, y_pred)
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title('Linear Regression Example')
plt.show()
```

In this example, we use the `fetch_california_housing()` function to load the California Housing dataset, which contains information on housing prices and other variables for different districts in California. We then extract the input and output data from the dataset and create a linear regression model using the `Scikit-learn` library's `LinearRegression()` function. We fit the model to the data using the `fit()` method.

Next, we use the model to generate predictions on the input data using the `predict()` method. Finally, we plot the actual versus predicted values using Matplotlib.

When we run this code, we get a plot that shows the actual housing prices on the x-axis and the predicted prices on the y-axis with regression trend line:

```python
from sklearn.datasets import fetch_california_housing
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Load the California Housing dataset
X, y = fetch_california_housing(return_X_y=True)

# Create a linear regression model and fit the data
model = LinearRegression().fit(X, y)

# Generate predictions on new data
y_pred = model.predict(X)

# Plot the actual and predicted values
plt.scatter(y, y_pred)
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title('Linear Regression Example')

# Add regression line to plot
plt.plot([y.min(), y.max()], [y.min(), y.max()], '--', color='red', label='Regression Line')
plt.legend()

plt.show()
```

A more readable form for dense data in the scatter plot is to use hexbins:

```python
from sklearn.datasets import fetch_california_housing
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np

# Load the California Housing dataset
X, y = fetch_california_housing(return_X_y=True)

# Create a linear regression model and fit the data
model = LinearRegression().fit(X, y)

# Generate predictions on new data
y_pred = model.predict(X)

# Plot the actual and predicted values
fig, ax = plt.subplots()
hb = ax.hexbin(y, y_pred, gridsize=50, cmap='Blues', edgecolors='grey')
cb = fig.colorbar(hb, ax=ax)

# Add regression line to plot
plt.plot([y.min(), y.max()], [y.min(), y.max()], '--', color='red', label='Regression Line')
plt.legend()

ax.set_xlabel('Actual')
ax.set_ylabel('Predicted')
ax.set_title('Linear Regression Example with California Housing dataset')
plt.show()
```

## Choosing the best evaluation metric for regression problems

A regression problem is a type of supervised machine learning problem that involves predicting a continuous numerical output variable based on one or more input variables. For example, if we want to predict the housing prices based on various features like the number of bedrooms and bathrooms, square footage, and location, we can use a regression model.

Now, when it comes to evaluating the performance of our regression model, we have various metrics at our disposal. The choice of evaluation metric depends on our goals and the nature of the problem we are trying to solve.

Let's take an example, suppose we choose a metric that significantly amplifies large errors and punishes them. In that case, even a single large error can have a significant impact on the overall error measure of our model. This can make our model appear to be performing poorly, even though it may be doing well in other aspects.

However, in some cases, a single large error may not be as important as many smaller errors. Therefore, it is important to choose a metric that places equal weight on all errors if our goal is to accurately predict housing prices, rather than minimizing the overall error.

To sum up, when dealing with regression problems, we need to carefully choose an appropriate evaluation metric that aligns well with our goals.
