## Intro - General concepts
### ANN
An Artificial Neural Network (ANN) is a machine learning model that is inspired by the structure and function of biological neurons in the brain. The basic building block of an ANN is a neuron or node, which receives inputs, performs calculations, and produces outputs.

Here is an example of how an ANN neuron works:

Inputs: Let's say we have an ANN that predicts whether an email is spam or not. The input layer of the network receives data from the email such as the sender, subject, and body.

Calculations: Each input is multiplied by a weight value, and these weighted inputs are then summed up. This sum is then passed through an activation function, which decides whether the neuron should fire or not. The activation function introduces non-linearity into the model, which allows it to learn complex relationships between the inputs and outputs.

Outputs: The output of the neuron is the result of the activation function. For example, if the activation function is a step function, the output would be 1 if the neuron fires and 0 if it does not.

Here's an example of the calculations involved:

Let's say the neuron receives inputs `x1`, `x2`, and `x3`, with corresponding weight values `w1`, `w2`, and `w3`. The sum of the weighted inputs is:
$$
sum = (x_1 * w_1) + (x_2 * w_2) + (x_3 * w_3)
$$
This sum is then passed through an activation function, such as the sigmoid function:
$$
output = \frac{1}{1 + exp(-sum)}
$$
The output will be a value between 0 and 1, representing the probability that the email is spam. If the output is above a certain threshold, say 0.5, the email will be classified as spam.

This is just a simplified example of how an ANN neuron works. In reality, ANNs can have many layers and thousands of neurons, with more complex activation functions and training algorithms.
### Bias
In an artificial neural network, bias is an additional input value that is added to the weighted sum of the inputs before passing it through the activation function. The bias is an important parameter in training the neural network, as it allows the model to adjust the decision boundary and make more accurate predictions.

Bias is represented as a constant value that is added to the weighted sum of the inputs. It is essentially an intercept term that determines the position of the activation function along the y-axis. By adjusting the value of the bias input, we can shift the decision boundary of the neuron and make it more or less likely to fire.

Mathematically, the calculation for the output of a neuron with bias is:
```
output = activation_function(weighted_sum + bias)
```
Here, the weighted_sum is the sum of the product of the input values and their respective weights. The bias input is added to this value before passing it through the activation function.

The bias term is learned during the training process of the neural network, along with the weights. During each iteration of the training process, the bias input is updated based on the error in the model's prediction. The goal is to find the optimal values for the weights and bias inputs that minimize the error and improve the accuracy of the model's predictions.

In summary, bias is an additional input value that is added to the weighted sum of the inputs in a neuron, and it allows the model to adjust the decision boundary and make more accurate predictions.

### Activation function introduces the non-linearity
Activation functions and non-linearity are closely related in the context of artificial neural networks.

An activation function is a mathematical function applied to the output of a neuron in a neural network, which introduces non-linearity into the model. Without an activation function, the model would be a linear function of the input data, and would not be able to represent complex non-linear relationships between the inputs and outputs.

Non-linearity refers to the property of a mathematical function where the output is not proportional to the input. In other words, non-linear functions do not follow a straight line when plotted on a graph. Activation functions are typically non-linear, which allows them to represent complex patterns and relationships in the data.

The sigmoid function is an S-shaped curve that produces an output between 0 and 1, while the ReLU function produces an output of 0 if the input is negative, and the input value itself if it is positive. Both of these functions are non-linear and introduce non-linearity into the model.

Activation functions are a way of introducing non-linearity into neural networks, which is essential for the model to learn complex patterns and relationships in the data. ### Activation functions

Activation functions are an essential component of artificial neural networks because they introduce non-linearity into the model, allowing it to learn and approximate complex relationships between the inputs and outputs. Without an activation function, the neural network would simply be a linear function of the inputs, which would severely limit its ability to model complex patterns and relationships in the data.

An activation function takes in the _weighted sum of the **inputs** and the **bias**_ term and produces an output that is passed on to the next layer of the network. The output of the activation function is what determines whether or not the neuron should fire, or activate, based on the inputs it receives.

Here's an example of how an activation function works:

Let's say we have a neuron that receives inputs `x1` and `x2`, with corresponding weights w1 and w2, and a bias term b. The weighted sum of the inputs is calculated as:

$$weighted\_sum = (x_1 * w_1) + (x_2 * w_2) + b$$

If we didn't have an activation function, the output of the neuron would simply be the weighted sum of the inputs. However, by applying an activation function, we can introduce non-linearity into the model and allow it to learn more complex patterns.

One commonly used activation function is the sigmoid function, which has an S-shaped curve and produces an output between 0 and 1. The sigmoid function is defined as:
$$ sigmoid(x) = \frac{1}{1 + exp(-x)} $$
So, the output of the neuron with the sigmoid activation function would be:
$$
output = sigmoid(weighted\_sum)
$$
Let's say we have the following values for the inputs, weights, and bias:
```
x1 = 2, x2 = 3, w1 = 0.5, w2 = 0.75, b = -1
```
The weighted sum of the inputs would be:
```
weighted_sum = (2 * 0.5) + (3 * 0.75) - 1
= 1.25
```
The output of the neuron with the sigmoid activation function would be:
```
output = sigmoid(weighted_sum)
= 1 / (1 + exp(-1.25))
= 0.777
```
So, the neuron would fire with an output of 0.777, indicating that the input data is more likely to belong to the positive class.

In summary, activation functions are necessary in artificial neural networks because they introduce non-linearity into the model and allow it to learn complex patterns and relationships in the data. By applying an activation function to the weighted sum of the inputs and the bias term, we can determine whether or not the neuron should fire, and produce an output that is passed on to the next layer of the network.

## What are ANNs?

<img src="https://drive.google.com/uc?export=view&id=1x8jKxtZ63wjom_Wb44OU-NM7vZWCeNVe" width=800 alt="img"/>


An ANN is made of many interconnected "**neurons**".   

<img src="https://drive.google.com/uc?export=view&id=1JEG7JoQ3qzbZunr1mUGDtf3GOTZ4NcnX" width=800 alt="img"/>

Each neuron takes in some floating point numbers (e.g. 1.0, 0.5, -1.0) and multiplies them by some other floating point numbers (e.g. 0.7, 0.6, 1.4) known as **weights** (1.0 * 0.7 = 0.7, 0.5 * 0.6 = 0.3, -1.0 * 1.4 = -1.4).  The weights act as a mechanism to focus on, or ignore, certain inputs.  The weighted inputs then get summed together (e.g. 0.7 + 0.3 + -1.4 = -0.4) along with a **bias** value (e.g. -0.4 + ** -0.1 ** = -0.5).  

The summed value (x) is now transformed into an output value (y) according to the neuron's **activation function**  (y = **f**(x)). 


e.g. -0.5 --> -0.05 if we use the **Leaky Rectified Linear Unit (Leaky ReLU)** activation function: y = f(x) = f(-0.5) = max(0.1*-0.5, -0.5) = max(-0.05, -0.5) = -0.05

In larger ANNs with many layers, the neuron's output value (e.g. -0.05) would become the input for another neuron.


<img src="https://drive.google.com/uc?export=view&id=13NSl3HFtGPJarJWOEvTQwlzb81Mdcamn" width=800 alt="img"/>


However, one of the first ANNs was known as the perceptron and it consisted of only a single neuron.  

<img src="https://drive.google.com/uc?export=view&id=1ICSdd_EYWn1lQDQiNH4pJ7NNVJ-L8Dlj" width=800 alt="img"/>


The output of the perceptron's neuron acts as the final prediction. 

<img src="https://drive.google.com/uc?export=view&id=1gm_4ACRckCenpIm1J7mskxd8ROum1zZi" width=800 alt="img"/>


This means that each neuron is a linear binary classifier all on its own (e.g. an output value >= 0 would be the blue class, but an output value < 0 would be the red class)

Lets code our own Perceptron


````
import numpy as np 

class Neuron:  
    def __init__(self, n_inputs, bias = 0., weights = None):  
        self.b = bias
        if weights: self.ws = np.array(weights)
        else: self.ws = np.random.rand(n_inputs)

    def _f(self, x): #activation function (here: leaky_relu)
        return max(x*.1, x)   

    def __call__(self, xs): #calculate the neuron's output: multiply the inputs with the weights and sum the values together, add the bias value,
                            # then transform the value via an activation function
        return self._f(xs @ self.ws + self.b) 
  

````

So why do we need so many neurons in an ANN if one will suffice?  Well, individual neurons are unable to classify non-linearly separable data because they can only ever learn a linear decision boundary.

There can be more than one hidden layer  to allow the neural net to learn more complex decision boundaries. ### Part 1: Regression

Now our ANN will be performing supervised machine learning, so we need to create a labelled training set.   Our training data will have inputs and expected output labels for those inputs.   Treating this as a regression problem, our ANN will learn to associate certain RGB colour values to specific regions of the picture,  making the training inputs 2 values (the x,y coordinates of each pixel) and the training outputs 3 values (the r,g,b colour values for each pixel).  

````
training_inputs,training_outputs = [],[]
for row,rgbs in enumerate(image_array):
  for column,rgb in enumerate(rgbs):
    training_inputs.append((row,column))
    r,g,b = rgb
    training_outputs.append((r/255,g/255,b/255))
````

now lets create our ANN (A multi-layered perceptron - also known as a fully-connected feed-forward network):  

<img src="https://drive.google.com/uc?export=view&id=1RxvA5e8VFom4NL3UO-1eZ1zb7wRKg1Il" width=600 alt="img"/>


it should have 2 neurons in the input layer (since there are 2 values to take in: x & y coordinates).  It should have 3 neurons in the output layer (since there are 3 values to learn: r, g, b).   (The number of hidden layers and the number of neurons in each hidden layer are two **hyperparameters** to experiment with, as well as the number of epochs we will train it for, the activation function, etc) - e.g. we'll use 10 hidden layers with 100 neurons in each (making this a deep neural network)

````
from sklearn.neural_network import MLPRegressor

ann = MLPRegressor(activation='relu', 
                   learning_rate = 'adaptive',
                   alpha=1e-5,
                   hidden_layer_sizes= tuple(100 for _ in range(10)))

ann.fit(training_inputs, training_outputs)                         
````

the trained network can now predict the normalised rgb values for any given coordinates.

````
ann.predict([[1,1]])
````

lets use the ANN to predict the RGB values for every coordinate and lets display the predicted RGB values for the entire image to see how well it 

````
predicted_outputs = ann.predict(training_inputs)

predicted_image_array = np.zeros_like(image_array)
i = 0
for row,rgbs in enumerate(predicted_image_array):
  for column in range(len(rgbs)):
    r,g,b = predicted_outputs[i]
    predicted_image_array[row][column] = [r*255,g*255,b*255]
    i += 1
Image.fromarray(predicted_image_array)
````

````
# Importing the necessary libraries
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
````

````
"""
Neuron class representing a single neuron
"""

class Neuron:
    def __init__(self, n_inputs, weights=None):
        self.weights = weights if weights else np.random.randn(n_inputs)
        self.value = None

    def __call__(self, inputs):
        self.value = np.dot(inputs, self.weights)
        return self.value
````

````
"""
NeutralNetwork class representing a neural network
"""

class NeuralNetwork:
    def __init__(self, n_inputs, n_hidden1, n_hidden2, n_output):
        self.hidden1 = [Neuron(n_inputs) for _ in range(n_hidden1)]
        self.hidden2 = [Neuron(n_hidden1) for _ in range(n_hidden2)]
        self.output = Neuron(n_hidden2, weights=None)

    def forward(self, inputs):
        hidden1_outputs = [neuron(inputs) for neuron in self.hidden1]
        hidden2_outputs = [neuron(hidden1_outputs) for neuron in self.hidden2]
        output = self.output(hidden2_outputs)
        return output
````

````
# Set the amount of input data, hidden layers and output data
n_inputs = 3
n_hidden1 = 4
n_hidden2 = 4
n_output = 1

# Creation of a network
network = NeuralNetwork(n_inputs, n_hidden1, n_hidden2, n_output)
````

````
# Example input data
inputs = np.array([4, 2, 6])

# Data transfer over the network
output = network.forward(inputs)
````

````
# Display of input layer, hidden layers and output layer values
print("Values\n" + (20 * "-") + "\n")

for i, value in enumerate(inputs):
    print(f"Input_{i+1}: {value}")

print("\n")

for i, neuron in enumerate(network.hidden1):
    print(f"Hidden1_{i+1}: {neuron.value}")

print("\n")

for i, neuron in enumerate(network.hidden2):
    print(f"Hidden2_{i+1}: {neuron.value}")

print("\n")

print("Output:", output)
````

````
# Creating a graph
G = nx.DiGraph()

# Adding nodes
G.add_nodes_from(['Input_'+str(i+1) for i in range(n_inputs)] +
                 [f'Hidden1_{i+1}' for i in range(n_hidden1)] +
                 [f'Hidden2_{i+1}' for i in range(n_hidden2)] + ['Output'])

# Adding edges
G.add_edges_from([(f'Input_{i+1}', f'Hidden1_{j+1}') for i in range(n_inputs) for j in range(n_hidden1)])
G.add_edges_from([(f'Hidden1_{i+1}', f'Hidden2_{j+1}') for i in range(n_hidden1) for j in range(n_hidden2)])
G.add_edges_from([(f'Hidden2_{i+1}', 'Output') for i in range(n_hidden2)])

# Setting the position of the nodes
pos = {}
for i in range(n_inputs):
    pos[f'Input_{i+1}'] = (0, i+1)
for i, node in enumerate(['Hidden1', 'Hidden2']):
    for j in range(n_hidden1 if i == 0 else n_hidden2):
        pos[f'{node}_{j+1}'] = (i+1, j+1)
pos['Output'] = (3, 2)

# Choosing node colors
colors = ['red'] * n_inputs + ['blue'] * (n_hidden1 + n_hidden2) + ['green']

# Drawing a graph
nx.draw_networkx_nodes(G, pos, node_color=colors, node_size=800, edgecolors='black')
nx.draw_networkx_edges(G, pos, edge_color='gray', arrowsize=20)
nx.draw_networkx_labels(G, pos, font_size=10, font_color='black', font_weight='bold', labels={node: '' for node in G.nodes})

# Labels under hidden layers
layer_labels = {'Hidden1': 'hidden layer 1', 'Hidden2': 'hidden layer 2'}
for layer, label in layer_labels.items():
    nodes = [node for node in G.nodes if node.startswith(layer)]
    x, y = zip(*[pos[node] for node in nodes])
    x_mean = sum(x) / len(x)
    y_mean = min(y) - 0.5
    plt.text(x_mean, y_mean, label, ha='center', va='center', fontweight='bold')

# Label for the input layer
input_label = 'input layer'
input_nodes = [node for node in G.nodes if node.startswith('Input')]
x_input, y_input = zip(*[pos[node] for node in input_nodes])
x_input_mean = sum(x_input) / len(x_input)
y_input_mean = min(y_input) - 0.5
plt.text(x_input_mean, y_input_mean, input_label, ha='center', va='center', fontweight='bold')

# Label for the output layer
output_label = 'output layer'
output_nodes = [node for node in G.nodes if node == 'Output']
x_output, y_output = zip(*[pos[node] for node in output_nodes])
x_output_mean = sum(x_output) / len(x_output)
y_output_mean = min(y_output) - 1.5
plt.text(x_output_mean, y_output_mean, output_label, ha='center', va='center', fontweight='bold')

# Disabling the display of axes on the plot
plt.axis('off')
plt.show()
````
