# 1. Introduction to Evolutionary Computation - simulate the nature
Evolutionary computation is a problem-solving approach inspired by the principles of natural selection and evolution. It aims to find optimal or near-optimal solutions to complex problems by mimicking the process of biological evolution. Understanding the concept of evolutionary computation is crucial, as it forms the foundation for genetic algorithms and other related techniques.

Evolutionary computation involves creating a population of potential solutions, representing them in a specific format, evaluating their fitness based on their performance, and iteratively applying genetic operators such as selection, crossover, and mutation to generate new generations of solutions. Over time, this iterative process leads to the improvement and refinement of solutions, ultimately converging towards an optimal or near-optimal solution.

The key idea behind evolutionary computation is to harness the power of natural selection, which is the process through which species evolve and adapt to their environment over generations. The concept of "survival of the fittest" lies at the core of both natural selection and evolutionary computation. In nature, individuals with traits that provide a higher fitness, or ability to survive and reproduce, are more likely to pass on their genetic material to the next generation. Similarly, in evolutionary computation, solutions with higher fitness are more likely to be selected for reproduction and have a higher chance of contributing to the next generation.

By simulating the mechanisms of natural selection, evolutionary computation explores the solution space, gradually improving the quality of solutions over time. The search process involves creating new candidate solutions through recombination and mutation, evaluating their fitness, and selecting the most promising ones for further evolution. This iterative process continues until a termination criterion is met, such as reaching a maximum number of generations or obtaining a satisfactory solution.

Evolutionary computation offers several advantages in problem-solving. It:

* can handle complex, multi-dimensional, and non-linear optimization problems where traditional algorithms struggle
* allows for the exploration of a large solution space, considering diverse alternatives, and can potentially find innovative and unconventional solutions

Understanding the concept of evolutionary computation provides us with a powerful tool to tackle real-world optimization problems and opens doors to various related algorithms and techniques, such as genetic algorithms, evolutionary programming, evolutionary strategies, and swarm intelligence.

## Binary encoding represents solutions as strings of 0s and 1s:

Binary encoding is a method used to represent solutions in genetic algorithms. It involves representing each solution as a string of 0s and 1s, where each bit in the string corresponds to a specific attribute or variable of the solution. The length of the string is determined by the number of variables or components in the solution. \\

In binary encoding, each bit in the string corresponds to a specific variable or component of the solution. By mapping the bits to their corresponding variables, the genetic algorithm can manipulate and evolve the solutions through crossover and mutation operations.
 > EXAMPLE:
     Let's consider a specific example with two variables: x and y, each represented by 3 bits. We have a binary string `101011` that represents a solution. To decode this binary string into its corresponding values for `x` and `y`, we split the string into two parts: `101` and `011`. The first part, `101`, corresponds to the value of `x`, and the second part, `011`, corresponds to the value of y.
     In this example, the decoded values for `x` and `y` would be: \\
     `x = 101 (in binary) = 5 (in decimal)` \\
`y = 011 (in binary) = 3 (in decimal)` \\
Therefore, the binary-encoded solution `101011` decodes to `x = 5` and `y = 3`.

By using binary encoding, genetic algorithms can efficiently manipulate and evolve solutions by applying genetic operators like crossover and mutation, working on the binary representation of the solutions.

### Example binary encoding:
This example illustrates binary encoding for the **Knapsack Problem**.

In this example, we have a knapsack problem with `5` items. Each item is represented as a tuple containing its `(name, value, weight)`. The goal is to maximize the total value of the selected items while not exceeding the weight capacity of the knapsack (`15` in this case).

The `binary_encoding` function converts a solution (represented as a list of binary values) into a binary string. The `decoding` function converts a binary string back into a list of selected items. The example solution `[1, 0, 1, 0, 1]` represents selecting the first, third, and fifth items.

After encoding and decoding, we calculate the total weight of the selected items and check if it exceeds the weight capacity. The output will indicate the binary encoding, the selected items, and whether the weight capacity is exceeded or not.

You can modify the example by adding more items or changing the weight capacity to explore different scenarios of the Knapsack Problem with binary encoding.

````python
# Define the items and their values and weights
items = [("Item1", 10, 5), ("Item2", 8, 3), ("Item3", 15, 8), ("Item4", 6, 2), ("Item5", 18, 9)]
weight_capacity = 15

# Binary encoding function
def binary_encoding(solution):
    binary_string = ""
    for item_selected in solution:
        if item_selected:
            binary_string += "1"
        else:
            binary_string += "0"
    return binary_string

# Decoding function
def decoding(binary_string):
    selected_items = []
    for i in range(len(binary_string)):
        if binary_string[i] == "1":
            selected_items.append(items[i][0])
    return selected_items

# Example solution
solution = [1, 0, 1, 0, 1]

# Encoding the solution
binary_string = binary_encoding(solution)
print("Binary Encoding: ", binary_string)

# Decoding the binary string
selected_items = decoding(binary_string)
print("Selected Items: ", selected_items)

# Check if the selected items exceed the weight capacity
total_weight = sum([items[i][2] for i in range(len(items)) if solution[i]])
if total_weight > weight_capacity:
    print("Weight capacity exceeded.")
else:
    print("Total Weight: ", total_weight)
````

````python
import numpy as np

# Rosenbrock function
def rosenbrock(x, y):
    return (1 - x)**2 + 100 * (y - x**2)**2

# Real-valued encoding function
def real_valued_encoding(solution):
    x = solution[0]
    y = solution[1]
    return x, y

# Decoding function
def decoding(encoded_solution):
    x, y = encoded_solution
    return x, y

# Example solution
solution = [2.5, -1.5]

# Encoding the solution
encoded_solution = real_valued_encoding(solution)
print("Encoded Solution: ", encoded_solution)

# Decoding the encoded solution
decoded_solution = decoding(encoded_solution)
print("Decoded Solution: ", decoded_solution)

# Evaluating the solution using the Rosenbrock function
fitness = rosenbrock(*decoded_solution)
print("Fitness: ", fitness)
````

## Permutation Encoding
Permutation encoding is a representation method used in genetic algorithms (GA) for problems that involve ordering or sequencing. It is particularly suitable for problems where the arrangement of elements is important, such as the Traveling Salesman Problem (TSP), job scheduling, and DNA sequence assembly.

In permutation encoding, a solution is represented as a permutation or ordering of elements. Each element in the permutation represents a unique item, city, task, or gene, depending on the problem domain. The position of an element in the permutation denotes its position in the sequence or order.

For example, consider the TSP where the goal is to find the shortest route that visits a set of cities exactly once and returns to the starting city. A permutation encoding can represent different city visitation orders. The position of each city in the permutation represents its visitation order in the route.

Let's say we have a TSP with four cities: `A`, `B`, `C`, and `D`. One possible permutation encoding could be `[A, C, D, B]`. This encoding suggests that the salesman starts at city `A`, then visits city `C`, followed by city `D`, and finally city `B` before returning to the starting city.

In the context of job scheduling, permutation encoding can be used to represent the order of tasks or jobs to be executed. Each position in the permutation corresponds to a specific task, and the value at that position denotes the identity or label of the task.

For example, let's consider a job scheduling problem with three tasks: `Task1`, `Task2`, and `Task3`. A permutation encoding `[Task2, Task3, Task1]`
indicates that `Task2` should be executed first, followed by `Task3`, and then `Task1`.

Permutation encoding allows genetic algorithms to explore the space of possible arrangements or sequences systematically. Through genetic operators like crossover and mutation, the GA can generate new permutations by combining or modifying existing ones. Fitness evaluation is performed based on the quality or objective function of the problem, which assesses the fitness or performance of each permutation.

In summary, permutation encoding provides a concise and effective way to represent ordering or sequencing problems in genetic algorithms. It allows GA to explore and optimize the arrangement of elements to find optimal or near-optimal solutions for problems like the Traveling Salesman Problem, job scheduling, DNA sequence assembly, and many other applications that involve sequencing or ordering elements.

### Example - permutation encoding
Illustrates permutation encoding for the Traveling Salesman Problem (TSP):

````python
import random

# List of cities with their coordinates
cities = {
    "City1": (0, 0),
    "City2": (1, 2),
    "City3": (3, 1),
    "City4": (2, 3),
    "City5": (4, 4)
}

# Calculate the distance between two cities using Euclidean distance
def calculate_distance(city1, city2):
    x1, y1 = cities[city1]
    x2, y2 = cities[city2]
    distance = ((x2 - x1) ** 2 + (y2 - y1) ** 2) ** 0.5
    return distance

# Permutation encoding function
def permutation_encoding(solution):
    encoded_solution = list(solution)
    random.shuffle(encoded_solution)
    return encoded_solution

# Decoding function
def decoding(encoded_solution):
    decoded_solution = list(encoded_solution)
    return decoded_solution

# Generate a random solution
solution = list(cities.keys())
random.shuffle(solution)

# Encoding the solution
encoded_solution = permutation_encoding(solution)
print("Encoded Solution: ", encoded_solution)

# Decoding the encoded solution
decoded_solution = decoding(encoded_solution)
print("Decoded Solution: ", decoded_solution)

# Calculate the total distance of the TSP route
total_distance = 0
for i in range(len(decoded_solution) - 1):
    city1 = decoded_solution[i]
    city2 = decoded_solution[i + 1]
    distance = calculate_distance(city1, city2)
    total_distance += distance

# Add the distance between the last and first city
first_city = decoded_solution[0]
last_city = decoded_solution[-1]
distance = calculate_distance(last_city, first_city)
total_distance += distance

print("Total Distance: ", total_distance)
````

In this example, we have a TSP with five cities represented by their names and coordinates. The goal is to find the shortest route that visits all cities exactly once and returns to the starting city.

The `calculate_distance` function calculates the Euclidean distance between two cities based on their coordinates.

The `permutation_encoding` function takes a solution (represented as a list of city names) and returns the same list as the encoded solution. In permutation encoding, the order of cities represents the visitation order.

The `decoding` function simply returns the encoded solution as the decoded solution.

We generate a random solution by shuffling the list of city names. The encoded solution is obtained by applying the permutation encoding function to the solution.

We then decode the encoded solution, which remains the same in this case since it was already in the desired format.

To calculate the total distance of the TSP route, we iterate through the decoded solution and calculate the distance between consecutive cities using the `calculate_distance` function. We also add the distance between the last city and the first city to complete the loop.

The output displays the encoded solution, the decoded solution (same as the original solution), and the total distance of the TSP route.

````python
#@title Visualization 
import matplotlib.pyplot as plt

# List of cities with their coordinates
cities = {
    "City1": (0, 0),
    "City2": (1, 2),
    "City3": (3, 1),
    "City4": (2, 3),
    "City5": (4, 4)
}

# Calculate the distance between two cities using Euclidean distance
def calculate_distance(city1, city2):
    x1, y1 = cities[city1]
    x2, y2 = cities[city2]
    distance = ((x2 - x1) ** 2 + (y2 - y1) ** 2) ** 0.5
    return distance

# Function to draw the TSP graph
def draw_tsp_graph(cities, route):
    fig, ax = plt.subplots()

    # Draw the cities
    for city, (x, y) in cities.items():
        ax.plot(x, y, 'ro')
        ax.text(x, y, city, ha='center', va='bottom')

    # Draw the route
    for i in range(len(route) - 1):
        city1 = route[i]
        city2 = route[i + 1]
        x1, y1 = cities[city1]
        x2, y2 = cities[city2]
        distance = calculate_distance(city1, city2)
        ax.arrow(x1, y1, x2 - x1, y2 - y1, length_includes_head=True, head_width=0.1, head_length=0.1, fc='k', ec='k')
        ax.text((x1 + x2) / 2, (y1 + y2) / 2, "{:.2f}".format(distance), ha='center', va='center')

    # Add the arrow and distance for the last city
    first_city = route[0]
    last_city = route[-1]
    x1, y1 = cities[last_city]
    x2, y2 = cities[first_city]
    distance = calculate_distance(last_city, first_city)
    ax.arrow(x1, y1, x2 - x1, y2 - y1, length_includes_head=True, head_width=0.1, head_length=0.1, fc='k', ec='k')
    ax.text((x1 + x2) / 2, (y1 + y2) / 2, "{:.2f}".format(distance), ha='center', va='center')

    plt.xlabel("X")
    plt.ylabel("Y")
    plt.title("Traveling Salesman Problem")
    plt.show()

# Example solution
solution = ["City1", "City3", "City5", "City2", "City4"]

# Draw the TSP graph
draw_tsp_graph(cities, solution)
````

````python
# Calculating probability for roulette wheel selection
beta = 1
# Create for list
costs = []

for i in range(len(population)):
   # list of all the population cost
   costs.append(population[i]['cost'])
costs = np.array(costs)
# take mean of all
avg_cost = np.mean(costs)
if avg_cost != 0:
    
    """We calculate each parent's probability by the exponential of negative beta times costs, 
    where beta is a pre-defined integer and costs is the cost of each parent divided 
    by the average cost of all the parents in the population."""

    costs = costs/avg_cost
probs = np.exp(-beta*costs)

def roulette_wheel_selection(p):
    # cumulative sum creating linear roulette wheel
    c = np.cumsum(p)
    # sum(p) = calculate to 100%
    # create random number
    r = sum(p) * np.random.rand()
    # sometimes r becomes bigger than c, drop them
    # choose non_zero population
    ind = np.argwhere(r <= c)
    # get the index of the first parent whose cumulative value is greater than the random number.
    return ind[0][0]

# roulette_wheel_selection(probs) choose randomly indices from population.
# lets create parents:

p1 = population[roulette_wheel_selection(probs)]
print(p1)
p2 = population[roulette_wheel_selection(probs)]
print(p2)
````

4. Crossover: \\

* Now that we got our two parents for breeding, the next step is to perform crossover/mating/breeding. Crossover refers to the process where certain genes from both the parent chromosomes are overlapped or mixed or swapped to produce new offspring. Since the offspring is the result of the parent chromosomes' crossover, it inherits both the parents' characteristics. There are three methods to perform crossover.

![](https://images.squarespace-cdn.com/content/v1/5c5aed8434c4e20e953d6011/1602410149373-1VJFUM1VLQYQF1O2HZ75/crossing+over.jpg?format=1000w)
Crossover involves combining genetic information from two selected parents to create new offspring. It promotes the exchange and recombination of genetic material, potentially generating novel and diverse solutions.

ðŸ“ **Algorithm:** Crossover

ðŸ”Ž **Input:** `parent1`, `parent2`

> _Crossover_(`parent1`, `parent2`)
   - Set `offspring` as an empty offspring individual
   - **For** each `gene` in `parent1` and `parent2`, do the following:
     - **If** random probability < `crossoverRate`, then do the following:
       - Transfer `gene` to `offspring`
   - **Return** `offspring`




 #### Single-point crossover: 
* In this method, both the parent chromosomes are cut at the same random point, and the leftover parts are swapped to produce two new offspring chromosomes. Yellow-colored genes represent the cutoff part of the chromosome.
![](https://miro.medium.com/max/1400/0*4yZ7VBKdQV-kcwBv.jpeg)

#### Two-point crossover: 
* A method similar to the single-point crossover, but the only difference is that the parent chromosomes are cut at two random points. Again, the yellow-colored cut off part of the chromosome is swapped.
![](https://miro.medium.com/max/1400/0*pQL21129SObz2vAR.jpeg)

#### Uniform crossover: 
* We first randomly choose which genes are supposed to be inherited from both the parent chromosomes and genes not inherited are marked in yellow color. Then, we model them as 0s and 1s, which are written in green color. The gene to be inherited is encoded as `1`, and the gene that should not be inherited is encoded as `0`. This series of `0`s and `1`s will be referred to as alpha from now on. Multiply the gene value with the corresponding alpha value for both the parents and then add the results to generate a single gene of the offspring chromosome. Let's consider the first gene of each parent chromosome. For parent-1, the gene value is `1`, and the corresponding alpha value is also `1`; hence, `1x1=1`. For parent-2, gene value is 0 and the corresponding alpha value is also 0 hence, `0x0=0`. The first gene of the offspring chromosome is `1+0=1`, and so on â€” this way, we get offspring-1, to reproduce the offspring-2, we take the compliment values of alpha and carry out the same process.
![](https://miro.medium.com/max/1400/0*dUczZqdp4q7N_UqU.jpeg)


Let $\mathbf{p}_1$ and $\mathbf{p}_2$ represent the parent individuals' genotypes. The crossover operator produces two offspring individuals, $\mathbf{o}_1$ and $\mathbf{o}_2$, by combining genetic material from the parents. The crossover can be represented as:

$$o_1,o_2=crossover(p_1,p_2)$$



* Programatically, we copy both the parents into the child variable: c1, c2. Randomly generate uniformly distributed alpha values between 0 and 1, which is the parent chromosome's shape (position). The rest of the process remains the same, except, in theory, we take the complement of alpha values to produce offspring-2, whereas, in the program, we swap the parents while multiplying with alpha, which is the same as taking the complement of alpha values.

````python
import copy
def crossover(p1, p2):
    # take a copy of parents
    c1 = copy.deepcopy(p1)
    c2 = copy.deepcopy(p2)
    alpha = np.random.uniform(0, 1, *(c1['position'].shape))
    c1['position'] = alpha*p1['position'] + (1-alpha)*p2['position']
    c2['position'] = alpha*p2['position'] + (1-alpha)*p1['position'] 
    return c1, c2
````

# 'Variation' in GA  leads to the exploration of the search space.
In the context of evolutionary algorithms, variation refers to the introduction of diversity or changes in the genetic makeup of individuals within the population. It is a crucial aspect that drives the exploration of the search space, allowing the algorithm to search for optimal or near-optimal solutions.

Variation occurs through two fundamental operators: crossover and mutation.

1. **Crossover** : Crossover involves combining genetic information from two parent individuals to create new offspring. It mimics the natural process of reproduction and recombination. During crossover, selected parts of the genetic material from the parents are exchanged or mixed, generating novel genetic combinations in the offspring.

The purpose of crossover is to explore the search space by creating offspring that inherit favorable traits from their parents. By combining genetic material from different individuals, crossover allows for the exploration of different genetic configurations that may lead to improved solutions. It promotes the sharing and recombination of good building blocks present in the population, increasing the chances of generating diverse and potentially better solutions.

2. **Mutation**: Mutation introduces random changes in the genetic material of individuals. It is a mechanism to introduce small perturbations or alterations to the genetic information, mimicking the process of genetic variation observed in natural evolution. Mutation is typically applied to one or more genes of an individual with a low probability.

The purpose of mutation is to introduce exploration and prevent the algorithm from getting stuck in local optima. It helps to maintain diversity within the population by introducing random changes that can potentially lead to entirely new solutions. Mutation enables the exploration of regions in the search space that might not be reachable through crossover alone. Even though mutation may initially degrade the fitness of an individual, it plays a crucial role in the long-term improvement of the population by diversifying the genetic makeup.

By combining crossover and mutation, evolutionary algorithms ensure a balance between exploration and exploitation. Crossover primarily focuses on exploiting good solutions within the current population, while mutation promotes exploration by introducing random changes. Together, these operators allow the algorithm to explore different regions of the search space, potentially discovering better solutions and avoiding premature convergence to suboptimal solutions.

Through variation, evolutionary algorithms maintain a diverse population that represents different regions of the search space. This diversity enables the algorithm to explore a wide range of potential solutions, adapt to changing environments, and ultimately converge towards optimal or near-optimal solutions for the given problem.

## Elitism
Elitism is a mechanism in genetic algorithms where a certain number of the best individuals from the current population are directly carried over to the next generation without undergoing any genetic operations (crossover and mutation). Elitism helps preserve the best solutions found so far and prevents the loss of valuable genetic material. By maintaining a small elite population, genetic diversity can be balanced with exploitation of the best solutions.

# Convergence criteria used to terminate evolutionary algorithms 

Convergence criteria in evolutionary algorithms determine when to terminate the algorithm based on specific conditions. The choice of convergence criteria depends on the problem at hand and the desired level of solution quality. Let's discuss two common convergence criteria: maximum generation limit and fitness threshold.

1. Maximum Generation Limit: \\
This convergence criterion terminates the algorithm after a predetermined number of generations. It assumes that after a certain number of iterations, the algorithm has had sufficient time to explore and converge towards a satisfactory solution.


2. Fitness Threshold: \\
This convergence criterion terminates the algorithm when the fitness value of the best individual in the population surpasses or meets a predefined threshold. It assumes that once a satisfactory fitness level is achieved, further iterations are unnecessary.
The fitness threshold convergence criterion can be represented as $f_{\text{best}} \geq \text{fitnessThreshold}$, where $f_{\text{best}}$ represents the fitness value of the best individual in the population.

These convergence criteria can be combined or used individually, depending on the problem requirements. The choice of convergence criteria should be based on the problem's characteristics and the desired stopping condition for the evolutionary algorithm.

#  The advantages of using evolutionary algorithms for optimization and problem-solving tasks.

Evolutionary algorithms offer several advantages for optimization and problem-solving tasks. Here are some key advantages:

1. Ability to Handle Complex Problems: Evolutionary algorithms excel at solving complex problems characterized by high-dimensional search spaces, non-linearity, and multiple objectives. They are particularly effective when traditional optimization techniques face challenges or become infeasible.

2. Global Search Capability: Evolutionary algorithms employ population-based search strategies that explore the solution space globally. By maintaining a diverse set of solutions, they can overcome local optima and converge towards globally optimal or near-optimal solutions.

3. Robustness and Adaptability: Evolutionary algorithms are robust in handling noisy and uncertain environments. # Implementing of basic evolution algorithm

## Example 1.
Here are code examples in Python to illustrate the implementation of basic evolutionary algorithms, including the initialization of a population, fitness evaluation, selection, crossover, and mutation operations. Stop condition is number of generations:

````python
import random

# Genetic Algorithm class
class GeneticAlgorithm:
    def __init__(self, population_size, chromosome_length, crossover_rate, mutation_rate):
        self.population_size = population_size
        self.chromosome_length = chromosome_length
        self.crossover_rate = crossover_rate
        self.mutation_rate = mutation_rate
        self.population = []

    # Initialize the population with random chromosomes
    def initialize_population(self):
        for _ in range(self.population_size):
            chromosome = [random.randint(0, 1) for _ in range(self.chromosome_length)]
            self.population.append(chromosome)

    # Evaluate the fitness of each individual in the population
    def evaluate_fitness(self):
        fitness_values = []
        for chromosome in self.population:
            fitness = sum(chromosome)  # Example fitness function (sum of ones)
            fitness_values.append(fitness)
        return fitness_values

    # Select individuals for reproduction (roulette wheel selection)
    def selection(self, fitness_values):
        selected_population = []
        total_fitness = sum(fitness_values)

        for _ in range(self.population_size):
            rand = random.uniform(0, total_fitness)
            cumulative_fitness = 0
            for i, fitness in enumerate(fitness_values):
                cumulative_fitness += fitness
                if cumulative_fitness >= rand:
                    selected_population.append(self.population[i])
                    break

        return selected_population

    # Perform crossover between two parent individuals
    def crossover(self, parent1, parent2):
        if random.random() < self.crossover_rate:
            crossover_point = random.randint(1, self.chromosome_length - 1)
            child1 = parent1[:crossover_point] + parent2[crossover_point:]
            child2 = parent2[:crossover_point] + parent1[crossover_point:]
            return child1, child2
        else:
            return parent1, parent2

    # Perform mutation on an individual
    def mutation(self, individual):
        mutated_individual = []
        for gene in individual:
            if random.random() < self.mutation_rate:
                mutated_individual.append(1 - gene)  # Flip the gene (0 to 1 or 1 to 0)
            else:
                mutated_individual.append(gene)
        return mutated_individual

    # Execute the evolutionary algorithm
    def evolve(self, num_generations):
        self.initialize_population()

        for generation in range(num_generations):
            fitness_values = self.evaluate_fitness()
            selected_population = self.selection(fitness_values)
            new_population = []

            while len(new_population) < self.population_size:
                parent1 = random.choice(selected_population)
                parent2 = random.choice(selected_population)
                child1, child2 = self.crossover(parent1, parent2)
                mutated_child1 = self.mutation(child1)
                mutated_child2 = self.mutation(child2)
                new_population.append(mutated_child1)
                new_population.append(mutated_child2)

            self.population = new_population

            # Printing the best individual of each generation
            best_fitness = max(fitness_values)
            best_individual = self.population[fitness_values.index(best_fitness)]
            print(f"Generation {generation + 1}: Best Fitness = {best_fitness}, Best Individual = {best_individual}")


# Example usage
population_size = 100
chromosome_length = 10
crossover_rate = 0.8
mutation_rate = 0.01
num_generations = 50

ga = GeneticAlgorithm(population_size, chromosome_length, crossover_rate, mutation_rate)
ga.evolve(num_generations)
````

In this application we'll start from a random image (random pixel values), then evolve the value of each pixel using the genetic algorithm.

The tricky part of this application is that an image is 2D or 3D, and the genetic algorithm expects the solutions to be 1D vectors. To tackle this issue we'll use the `img2chromosome()` function defined below to convert an image to a 1D vector.

````python
import functools
import operator

import numpy as np

def img2chromosome(img_arr):
    """
    Represents the image as a 1D vector.
    
    img_arr: The image to be converted into a vector.
    
    Returns the vector.
    """

    return np.reshape(a=img_arr, newshape=(functools.reduce(operator.mul, img_arr.shape)))
````

````python

#  restore the 2D or 3D image back from the vector
def chromosome2img(vector, shape):
    """
    Converts a 1D vector into an array.
    
    vector: The vector to be converted into an array.
    shape: The shape of the target array.
    
    Returns the array.
    """

    # Check if the vector can be reshaped according to the specified shape.
    if len(vector) != functools.reduce(operator.mul, shape):
        raise ValueError(f"A vector of length {len(vector)} into an array of shape {shape}.")

    return np.reshape(a=vector, newshape=shape)
````
