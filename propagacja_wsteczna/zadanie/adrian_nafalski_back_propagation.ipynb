# Intro

Artificial neural networks (ANNs) are composed of interconnected processing nodes that are organized into layers. The basic building blocks of ANNs are input layers, hidden layers, and output layers. The input layer receives the raw data, and the output layer produces the final predictions. The hidden layers are responsible for transforming the input data into a useful representation that can be used for prediction.

The connections between the nodes in an ANN are represented by weights, which determine the strength of the signal transmitted between nodes. The weights are initialized randomly and adjusted during the learning process to improve the accuracy of the predictions.

Backpropagation is a commonly used algorithm for training ANNs. It works by calculating the error between the predicted output and the actual output, and then propagating the error back through the network to adjust the weights. The goal of backpropagation is to minimize the difference between the predicted output and the actual output by adjusting the weights to make the predicted output more accurate.

The backpropagation algorithm involves several steps:

1. Forward propagation: The input data is fed through the network, and the output is calculated. Each node in the network receives input from the previous layer and applies a nonlinear activation function to produce an output.

2. Error calculation: The error between the predicted output and the actual output is calculated using a loss function, such as mean squared error or cross-entropy loss.

3.  Backward propagation: The error is propagated backwards through the network, and the gradient of the loss function with respect to the weights is calculated. This gradient is used to adjust the weights in the network to minimize the loss function.

4.  Weight update: The weights are updated using an optimization algorithm, such as stochastic gradient descent, that adjusts the weights in the direction that reduces the loss function.

This process is repeated for multiple epochs until the network converges to a set of weights that produces accurate predictions on the training data.

In addition to backpropagation, other techniques such as regularization and dropout are often used to improve the performance of ANNs and prevent overfitting.

# TL;DR

## What is backpropagation?

Backpropagation is a process involved in training a neural network. It involves taking the error rate of a forward propagation and feeding this loss backward through the neural network layers to fine-tune the weights. ²³ Backpropagation is the essence of neural net training. It is the method of fine-tuning the weights of a neural network based on the error rate obtained in the previous epoch (i.e., iteration). ³ Proper tuning of the weights allows you to reduce error rates and make the model reliable by increasing its generalization. ²

## How does backpropagation work?

Backpropagation works by applying the chain rule of calculus to compute the gradients of the loss function with respect to the weights of the network. The gradients indicate how much each weight contributes to the error, and how it should be adjusted to minimize the error. The gradients are then used to update the weights using an optimization algorithm, such as gradient descent.

The following steps describe how backpropagation works in a simple feedforward neural network with one hidden layer:

1. Initialize the weights randomly and set the learning rate.
2. Perform a forward pass to compute the output and the error for each training example.
3. Perform a backward pass to compute the gradients for each weight using the chain rule.
4. Update each weight by subtracting the product of the learning rate and the gradient from the current weight value.
5. Repeat steps 2-4 until the error is sufficiently small or a maximum number of iterations is reached.

## What are some formulas for backpropagation?

The formulas for backpropagation depend on the choice of activation functions and loss functions for the network. For simplicity, let's assume we use an identity activation function ($f(a) = a$) and a mean squared error loss function $(L(y,y') = (y-y')^2/2)$ for a regression problem.



Let's denote the input layer as layer 0, the hidden layer as layer 1, and the output layer as layer 2. Let's also denote the number of units in each layer as $n_0, n_1$, and $n_2$, respectively. Let's use $x(i)$ to denote the i-th input vector, $y(i)$ to denote the i-th target output scalar, and $y'(i)$ to denote the i-th predicted output scalar.

The formulas for computing the output and the error for each training example are:

$$y'(i) = \sum_{j=0}^{n_1} w_{2j} f(\sum_{k=0}^{n_0} w_{1jk} x(i)_k + b_{1j}) + b_2$$

$$e(i) = L(y(i),y'(i)) = \frac{(y(i)-y'(i))^2}{2}$$

The formulas for computing the gradients for each weight are:

$$\frac{\partial e(i)}{\partial w_{2j}} = \frac{\partial e(i)}{\partial y'(i)} \frac{\partial y'(i)}{\partial w_{2j}} = -(y(i)-y'(i)) f(\sum_{k=0}^{n_0} w_{1jk} x(i)_k + b_{1j})$$

$$\frac{\partial e(i)}{\partial w_{1jk}} = \frac{\partial e(i)}{\partial y'(i)} \frac{\partial y'(i)}{\partial w_{1jk}} = -(y(i)-y'(i)) w_{2j} x(i)_k$$

The formulas for updating each weight are:

$$w_{2j} := w_{2j} - \alpha \frac{1}{m} \sum_{i=1}^{m} \frac{\partial e(i)}{\partial w_{2j}}$$

$$w_{1jk} := w_{1jk} - \alpha \frac{1}{m} \sum_{i=1}^{m} \frac{\partial e(i)}{\partial w_{1jk}}$$

where m is the number of training examples and alpha is the learning rate.


# Gradient Descent - for function
Gradient descent is an optimization algorithm commonly used in machine learning to find the minimum of a function. The algorithm works by iteratively adjusting the parameters of a model in the direction of steepest descent of the objective function.

Here's a mathematical explanation of the gradient descent algorithm:

Suppose we have a function f(x), where x is a vector of parameters we want to optimize. The goal is to find the values of x that minimize the function f(x). The gradient descent algorithm iteratively updates the value of x based on the gradient of f(x) at the current value of x.

The gradient of a function is a vector that points in the direction of the steepest increase of the function at a given point. In other words, the gradient points in the direction of the largest increase in the function.

The update rule for gradient descent is:
$$
x = x - \alpha\nabla f(x)
$$
where:

* $x$ is the vector of parameters we are trying to optimize
* $α$ (alpha) is the learning rate, which controls the step size of the updates
* $\nabla f(x)$ is the gradient of the function f(x) at the current value of x

The gradient ∇f(x) is a vector that points in the direction of the steepest increase of f(x) at the current value of x. By subtracting this vector from x, we move in the opposite direction, towards the minimum of the function.

The learning rate α determines the step size of the updates. If α is too small, the algorithm will take a long time to converge to the minimum. If α is too large, the algorithm may overshoot the minimum and oscillate around it.

The gradient descent algorithm is usually applied in a batch mode, where the gradient is computed over a subset of the data at each iteration. This is known as batch gradient descent. Alternatively, the algorithm can be applied on a single data point at a time, which is known as stochastic gradient descent.

In summary, gradient descent is an optimization algorithm that iteratively updates the values of the parameters of a model in the direction of steepest descent of the objective function. The algorithm is controlled by the learning rate α, which determines the step size of the updates. The update rule is given by x = x - α∇f(x), where ∇f(x) is the gradient of the function f(x) at the current value of x.

## Gradient descent in context of Artificial Neural Networks 
In the context of neural networks, the gradient descent algorithm is used to optimize the weights and biases of the network in order to minimize a cost function.

#### Cost Function
The cost function is a measure of how well the network is performing on a particular task. It quantifies the difference between the predicted output of the network and the actual output, given a set of input data. The goal of the optimization process is to find the values of the weights and biases that minimize the cost function.

The cost function is an essential component of the training process of an ANN because it enables the optimization of the network's parameters (weights and biases) based on the minimization of the cost function.

The objective of the training process is to adjust the network's parameters in such a way that the cost function is minimized, meaning that the predicted output is as close as possible to the actual output. This process is often referred to as "minimizing the error" or "minimizing the loss."

There are various types of cost functions used in ANNs, and the choice of cost function depends on the specific task at hand. Some common cost functions include the mean squared error (MSE) function, which is often used for regression tasks, and the cross-entropy function, which is commonly used for classification tasks.

In summary, the cost function J is a measure of the error between the predicted output of an ANN and the actual output for a given set of input data. It is used in the training process to adjust the network's parameters in such a way that the cost function is minimized.

### Examplary Cost Functions
The cost function for a neural network can take many forms, depending on the specific task and architecture of the network. One commonly used cost function is the mean squared error (MSE), which measures the average squared difference between the predicted output and the actual output.The mean squared error (MSE) function is commonly used in regression tasks. The formula for the MSE function is:

$$
J(\theta) = \frac{1}{2m} \sum_{i=1}^{m} \left( h_{\theta}(x^{(i)}) - y^{(i)} \right)^2
$$
where:

* $J(\theta)$ is the MSE function
* $\theta$ are the weights of the regression model
* $m$ is the number of training examples
* $y^{(i)}$ is the actual label of the $i$-th training example
* $h_{\theta}(x^{(i)})$ is the predicted output of the regression model for the $i$-th training example


**Logistic Loss Function (aka binary cross-entropy loss function):**

The logistic loss function, also known as the binary cross-entropy loss function, is commonly used in binary classification tasks. The formula for the logistic loss function is:

$$
J(\theta) = -\frac{1}{m} \sum_{i=1}^{m} \left[ y^{(i)} \log h_{\theta}(x^{(i)}) + (1 - y^{(i)}) \log (1 - h_{\theta}(x^{(i)})) \right]
$$
where:

* $J(\theta)$ is the logistic loss function
* $\theta$ are the weights of the logistic regression model
* $m$ is the number of training examples
* $y^{(i)}$ is the actual label of the $i$-th training example
* $h_{\theta}(x^{(i)})$ is the predicted probability that the $i$-th training example belongs to the positive class, given the model parameters $\theta$
* $\log$ is the natural logarithm

**Cross-Entropy Function:**

The cross-entropy function is commonly used in multi-class classification tasks. The formula for the cross-entropy function is:
$$
J(\theta) = -\frac{1}{m} \sum_{i=1}^{m} \sum_{j=1}^{k} y_j^{(i)} \log h_{\theta}(x^{(i)})_j
$$

where:

* $J(\theta)$ is the cross-entropy function
* $\theta$ are the weights of the classification model
* $m$ is the number of training examples
* $y_j^{(i)}$ is the indicator variable that is equal to 1 if the $i$-th training example belongs to the $j$-th class, and 0 otherwise
* $h_{\theta}(x^{(i)})_j$ is the predicted probability that the $i$-th training example belongs to the $j$-th class, given the model parameters $\theta$
* $k$ is the number of classes


#### Minimize cost function $J$
The goal of the optimization process is to find the values of w and b that minimize the cost function J. This is done by computing the gradient of the cost function with respect to the weights and biases, and using this gradient to update the weights and biases in the direction of steepest descent.

The update rule for the weights and biases in a neural network is similar to the one used in the general gradient descent algorithm:
$$
\begin{align}
w &:= w - \alpha \frac{\partial J(w,b)}{\partial w} \\
b &:= b - \alpha \frac{\partial J(w,b)}{\partial b}
\end{align}
$$
where:

* $\alpha$ (alpha) is the learning rate, which controls the step size of the updates
* $\frac{\partial J(w,b)}{\partial w}$ and $\frac{\partial J(w,b)}{\partial b}$ are the partial derivatives of the cost function with respect to the weights and biases, respectively.

The learning rate $\alpha$ determines the step size of the weight and bias updates. If $\alpha$ is too small, the algorithm will take a long time to converge to the minimum. If $\alpha$ is too large, the algorithm may overshoot the minimum and oscillate around it.

In practice, different values of the learning rate may need to be tested to find the optimal one for a given problem. Additionally, other variations of gradient descent may be used, such as mini-batch or stochastic gradient descent, which use subsets of the training data at each iteration to compute the gradient.

Overall, the gradient descent algorithm is a powerful optimization technique that can be used to train neural networks by minimizing a cost function. The key components of the algorithm are the cost function, the learning rate, and the update rule for the weights and biases of the network.


## Gradient descent types:

There are 3 main GD types based on which part of input data we use to update weights:

* **Stochastic gradient descent**: Update network weights using one data point at a time.

* **Mini-batch gradient descent**: Update network weights using a few data points at a time.

* **Batch gradient descent**: Update network weights using the entire data at once.

For function $f(x)=x^2$ let's see how following steps of gradient descent closes to minimum of the function $f$

````python
#@title Gradient descent steps for $f(x)=x^2$
import numpy as np
import matplotlib.pyplot as plt

# Define the cost function
def cost_function(x):
    return x ** 2

# Define the gradient of the cost function
def gradient(x):
    return 2 * x

# Define the gradient descent algorithm
def gradient_descent(learning_rate, num_iterations):
    x = 10
    x_history = [x]
    cost_history = [cost_function(x)]
    for i in range(num_iterations):
        gradient_x = gradient(x)
        x -= learning_rate * gradient_x
        x_history.append(x)
        cost_history.append(cost_function(x))
    return x_history, cost_history

# Run the gradient descent algorithm
learning_rate = 0.1
num_iterations = 10
x_history, cost_history = gradient_descent(learning_rate, num_iterations)

# Plot the cost function and the steps of gradient descent
x_range = np.linspace(-10, 10, 100)
cost_range = cost_function(x_range)
fig, ax = plt.subplots()
ax.plot(x_range, cost_range, label='Cost function')
ax.scatter(x_history, cost_history, c='r', label='Gradient descent steps')
ax.set_xlabel('x')
ax.set_ylabel('Cost')
ax.set_title('Gradient descent for a simple cost function')

# Add coordinates next to each step and number the steps
for i, xy in enumerate(zip(x_history, cost_history)):
    ax.annotate(f'{i}', xy=xy, textcoords='offset points', xytext=(0,10), ha='center')
    ax.annotate(f'({xy[0]:.2f},{xy[1]:.2f})', xy=xy, textcoords='offset points', xytext=(0,-20), ha='center')

# Add coordinate numbers on axes
xticks = np.arange(-5, 6, 1)
yticks = np.arange(0, 26, 5)
ax.set_xticks(xticks)
ax.set_yticks(yticks)
ax.set_xticklabels([f'{x}' for x in xticks])
ax.set_yticklabels([f'{y}' for y in yticks])

plt.legend()
plt.show()
````

````python
#@title Contour Plot
import numpy as np
import matplotlib.pyplot as plt

def the_function_f(x):
    return x[0]**2 + x[1]**2

def grad_of_f(x):
    return np.array([2*x[0], 2*x[1]])

x0 = np.array([3, 4]) # initial point
alpha = 0.1 # learning rate
n_iter = 20 # number of iterations


x = x0 # current point
evolution_X1 = [] # list to store x1 values
evolution_X2 = [] # list to store x2 values

# We can then implement the gradient descent algorithm using a loop:
for i in range(n_iter):
    evolution_X1.append(x[0]) # append current x1 value
    evolution_X2.append(x[1]) # append current x2 value
    x = x - alpha * grad_of_f(x) # update x using gradient descent formula

# We can then plot the evolution of the points along with the level curves of the function:
# create a grid of points for plotting
X1, X2 = np.meshgrid(np.linspace(-5, 5, 100), np.linspace(-5, 5, 100))
Z = the_function_f([X1, X2])

# plot the level curves and the points
plt.figure(figsize=(8, 6))
plt.contour(X1, X2, Z, levels=np.logspace(0, 3, 10)) # plot level curves using log scale
plt.title("Evolution of the cost function during gradient descent with level circles", fontsize=15)
plt.plot(evolution_X1, evolution_X2) # plot line connecting the points
plt.plot(evolution_X1, evolution_X2, '*', label="Steps of GD on the function") # plot points with star marker
plt.xlabel('x1', fontsize=11)
plt.ylabel('x2', fontsize=11)
plt.legend()
plt.show()
````

````python
#@title Gradient descent steps for exemplary convex function: $f(x,y)=x^2+y^2$

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the cost function
def cost_function(x, y):
    return x ** 2 + y ** 2

# Define the gradient of the cost function
def gradient(x, y):
    return np.array([2 * x, 2 * y])

# Define the gradient descent algorithm
def gradient_descent(learning_rate, num_iterations):
    x = np.array([3.0, 3.0])
    x_history = [x]
    cost_history = [cost_function(*x)]
    for i in range(num_iterations):
        gradient_x, gradient_y = gradient(*x)
        x -= learning_rate * np.array([gradient_x, gradient_y])
        x_history.append(x)
        cost_history.append(cost_function(*x))
    return np.array(x_history), np.array(cost_history)

# Run the gradient descent algorithm
learning_rate = 0.1
num_iterations = 20
x_history, cost_history = gradient_descent(learning_rate, num_iterations)

# Generate a 3D surface plot of the cost function
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
x_range = np.linspace(-5, 5, 100)
y_range = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x_range, y_range)
Z = cost_function(X, Y)
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Cost')

# Add the path of gradient descent to the plot
ax.plot(x_history[:, 0], x_history[:, 1], cost_history, c='r', marker='o')

plt.show()
````

````python
#@title Gradient descent steps for finding local minima of exemplary NON-convex function: $$f(x)=sin(x)+cos(y)$$
import numpy as np
import plotly.graph_objs as go

# Define the function to optimize
def f(x, y):
    return np.sin(x) + np.cos(y)

# Define the partial derivatives
def partial_x(x, y):
    return np.cos(x)

def partial_y(x, y):
    return -np.sin(y)

# Set the learning rate
alpha = 0.1

# Set the initial point
x, y = 0.5, 0.5

# Initialize the history lists
x_history = [x]
y_history = [y]
cost_history = [f(x, y)]

# Set the stopping condition threshold
threshold = 1e-6

# Run gradient descent until convergence
while True:
    # Calculate the gradients
    dx = partial_x(x, y)
    dy = partial_y(x, y)
    
    # Update the parameters
    x -= alpha * dx
    y -= alpha * dy
    
    # Calculate the new cost
    cost = f(x, y)
    
    # Add the new values to the history lists
    x_history.append(x)
    y_history.append(y)
    cost_history.append(cost)
    
    # Check if the difference between the current and previous cost is below the threshold
    if len(cost_history) > 1 and abs(cost_history[-1] - cost_history[-2]) < threshold:
        break

# Create a 3D surface plot of the function
x_vals = np.linspace(-2*np.pi, 2*np.pi, 100)
y_vals = np.linspace(-2*np.pi, 2*np.pi, 100)
x_mesh, y_mesh = np.meshgrid(x_vals, y_vals)
z_mesh = f(x_mesh, y_mesh)

fig = go.Figure(data=[go.Surface(x=x_mesh, y=y_mesh, z=z_mesh)])

# Add the trace of the gradient descent path
fig.add_trace(go.Scatter3d(
    x=x_history,
    y=y_history,
    z=cost_history,
    mode='lines+markers',
    line=dict(color='red', width=3),
    marker=dict(size=5, color='black')
))

# Set the axis labels
fig.update_layout(scene=dict(
                    xaxis_title='X',
                    yaxis_title='Y',
                    zaxis_title='Cost'))

# Set the figure size
fig.update_layout(width=800, height=1200)

# Make the plot interactive
fig.show()
````

### Error calculation
After calculating the output of the neural network using forward propagation, the next step (in SUPERVISED LEARNING) is to calculate the error between the predicted output and the true output. This error can then be used to adjust the weights and biases of the network using backpropagation.

To calculate the error, we first need to define the true output of the network for the given input. For the sake of simplicity, let's assume that the true output is `[0.3, 0.7]`.

````python
# Define true output
y_true = np.array([0.3, 0.7])

# Calculate error
error = y_true - a3

print(error)
````

In this code, we define the true output as a numpy array. We then calculate the error as the element-wise difference between the true output and the predicted output `a3`. This gives us a numpy array containing the errors for each output neuron.




## Backpropagation
Now that we have the error, we can use backpropagation to adjust the weights and biases of the network in order to reduce the error. This involves calculating the gradient of the error with respect to each weight and bias in the network, and using these gradients to update the weights and biases in the opposite direction of the gradient.

Here are the 5 steps of backpropagation with the formulas

1. Compute the output layer error: $$\delta^{(3)} = (y - \hat{y}\ ) \odot \sigma'(\mathbf{z}^{(3)}) $$
2. Compute the error of the second hidden layer:
$$ \delta^{(2)} = (\delta^{(3)} \cdot \mathbf{W}^{(3)}) \odot \sigma'(\mathbf{z}^{(2)})
$$
3.    Compute the error of the first hidden layer:
$$
\delta^{(1)} = (\delta^{(2)} \cdot \mathbf{W}^{(2)}) \odot \sigma'(\mathbf{z}^{(1)})
$$
4. Compute the gradients with respect to the weights:
$$
\frac{\partial J}{\partial \mathbf{W}^{(3)}} = \mathbf{a}^{(2)}{}^\top \cdot \delta^{(3)} 
$$ $$
\frac{\partial J}{\partial \mathbf{W}^{(2)}} = \mathbf{a}^{(1)}{}^\top \cdot \delta^{(2)}
$$ $$
\frac{\partial J}{\partial \mathbf{W}^{(1)}} = \mathbf{x}{}^\top \cdot \delta^{(1)}
$$
5.     Compute the gradients with respect to the biases:
$$
\frac{\partial J}{\partial \mathbf{b}^{(3)}} = \sum_{i=1}^m \delta^{(3)}_i
$$ $$
\frac{\partial J}{\partial \mathbf{b}^{(2)}} = \sum_{i=1}^m \delta^{(2)}_i
$$ $$
\frac{\partial J}{\partial \mathbf{b}^{(1)}} = \sum_{i=1}^m \delta^{(1)}_i
$$

In the above formulas:
* $y$ is the true output,
* $\hat{y}$ is the predicted output,
* $\mathbf{z}^{(i)}$ is the weighted sum of the inputs to layer $i$,
* $\sigma(\cdot)$ is the sigmoid activation function,
* $\odot$ is the element-wise multiplication operator,
* $\mathbf{W}^{(i)}$ is the weight matrix for layer $i$,
* $\mathbf{a}^{(i)}$ is the activation output for layer $i$,
* $\mathbf{x}$ is the input data, 
* $\mathbf{b}^{(i)}$ is the bias vector for layer $i$, and $m$ is the number of training examples.


So backpropagation is about calculating the gradients of the loss function with respect to the weights and biases of the network. These gradients are then used to update the weights and biases using gradient descent.

Let's implement backpropagation for our example:

```Python
# Backpropagation with gradient descent

delta3 = error * sigmoid_prime(z3)
dJ_dW3 = a2.T.dot(delta3)

delta2 = delta3.dot(W3.T) * sigmoid_prime(z2)
dJ_dW2 = a1.T.dot(delta2)

delta1 = delta2.dot(W2.T) * sigmoid_prime(z1)
dJ_dW1 = X.T.dot(delta1)

# Update weights and biases
alpha = 0.1 # learning rate
W1 += alpha * dJ_dW1
b1 += alpha * np.sum(delta1, axis=0)

W2 += alpha * dJ_dW2
b2 += alpha * np.sum(delta2, axis=0)

W3 += alpha * dJ_dW3
b3 += alpha * np.sum(delta3, axis=0)
```

Let's go through each step of the code:

* We first calculate the delta of the output layer `delta3` as the element-wise product of the error and the derivative of the sigmoid function at `z3`, which we previously defined.

* We then calculate the gradient of the loss with respect to the weights of the output layer `dJ_dW3` using the transpose of the activation of the second hidden layer `a2` multiplied by delta3.

* We calculate the delta of the second hidden layer `delta2` as the element-wise product of `delta3` dot producted with the weight `W3`, and the derivative of the sigmoid function at `z2`.

* We calculate the gradient of the loss with respect to the weights of the second hidden layer `dJ_dW2` using the transpose of the activation of the first hidden layer a1 multiplied by `delta2`.

* We calculate the delta of the first hidden layer `delta1` as the element-wise product of `delta2` dot producted with the weight `W2`, and the derivative of the sigmoid function at `z1`.

* We calculate the gradient of the loss with respect to the weights of the first hidden layer `dJ_dW1` using the transpose of the input `X` multiplied by `delta1`.

* We then update the weights and biases of each layer using the learning rate `alpha` multiplied by the respective gradients. We also update the biases of each layer by summing up the deltas along the rows (`axis=0`).

By updating the weights and biases using the gradients calculated by backpropagation, we gradually reduce the error of the neural network and improve its performance.

# ANN Summary


## I. The Perceptron Model
We will start by discussing the perceptron model, which is the simplest form of an artificial neural network. The perceptron has one or more inputs, a bias, and a single output. The output of the perceptron is a binary value, which is determined by applying a step function to the weighted sum of the inputs and bias.

Code Example:

````python
def step_function(x):
    return 1 if x > 0 else 0

class Perceptron:
    def __init__(self, num_inputs, weights=None, bias=0):
        self.num_inputs = num_inputs
        self.weights = weights or [0] * num_inputs
        self.bias = bias
    
    def __call__(self, inputs):
        weighted_sum = sum(w * x for w, x in zip(self.weights, inputs)) + self.bias
        return step_function(weighted_sum)
````

## II. Multilayer Perceptrons
We will then move on to multilayer perceptrons (MLPs), which are ANNs with one or more hidden layers. MLPs can perform more complex tasks than perceptrons, as they are able to learn nonlinear decision boundaries. Each neuron in an MLP has an activation function, which is used to determine its output based on its input.

Code Example:

````python
class MLP:
    def __init__(self, num_inputs, hidden_layers, num_outputs):
        self.num_inputs = num_inputs
        self.hidden_layers = hidden_layers
        self.num_outputs = num_outputs
        
        # Initialize weights
        weights = []
        for i in range(len(hidden_layers) + 1):
            if i == 0:
                input_size = num_inputs
            else:
                input_size = hidden_layers[i-1]
            if i == len(hidden_layers):
                output_size = num_outputs
            else:
                output_size = hidden_layers[i]
            weights.append(np.random.randn(input_size, output_size))
        self.weights = weights
    
    def forward(self, inputs):
        activations = inputs
        for weight in self.weights:
            # Compute weighted sum
            hidden_inputs = np.dot(activations, weight)
            # Apply activation function
            activations = sigmoid(hidden_inputs)
        return activations

def sigmoid(x):
    return 1 / (1 + np.exp(-x))
````

## III. Forward Propagation
We will now move on to forward propagation, which is the process of computing the output of an MLP given some input. This involves computing the weighted sum of the inputs to each neuron, and applying the activation function to get the output of that neuron. This process is repeated for each layer until the final output is obtained.

Code Example:

````python
def feedforward(X, W, b):
    # input layer
    a1 = X
    
    # hidden layer 1
    z2 = np.dot(a1, W[0]) + b[0]
    a2 = sigmoid(z2)
    
    # hidden layer 2
    z3 = np.dot(a2, W[1]) + b[1]
    a3 = sigmoid(z3)
    
    # output layer
    z4 = np.dot(a3, W[2]) + b[2]
    a4 = sigmoid(z4)
    
    return (a1, z2, a2, z3, a3, z4, a4)
````

## IV. Backpropagation
$$
W^{(l)} = W^{(l)} - \alpha \cdot \nabla_{W^{(l)}} J(W,b,X,Y) \\
b^{(l)} = b^{(l)} - \alpha \cdot \nabla_{b^{(l)}} J(W,b,X,Y)
$$
where $\alpha$ is the learning rate, $\nabla_{W^{(l)}} J(W,b,X,Y)$ is the gradient of the loss function with respect to the weights of layer $l$, and $\nabla_{b^{(l)}} J(W,b,X,Y)$ is the gradient of the loss function with respect to the biases of layer $l$.

We repeat the above steps (2-4) for multiple epochs until the model converges to an acceptable level of accuracy.

Here is the updated Python code with backpropagation:

````python
def sigmoid_prime(z):
    return sigmoid(z) * (1 - sigmoid(z))

def backpropagation(X, Y, W, b, learning_rate):
    m = X.shape[0]
    
    # forward pass
    a1, z2, a2, z3, a3, z4, a4 = feedforward(X, W, b)
    
    # error calculation
    delta4 = (a4 - Y) * sigmoid_prime(z4)
    delta3 = np.dot(delta4, W[2].T) * sigmoid_prime(z3)
    delta2 = np.dot(delta3, W[1].T) * sigmoid_prime(z2)
    
    # gradient calculation
    dW2 = np.dot(a3.T, delta4) / m
    db2 = np.sum(delta4, axis=0, keepdims=True) / m
    dW1 = np.dot(a2.T, delta3) / m
    db1 = np.sum(delta3, axis=0) / m
    dW0 = np.dot(a1.T, delta2) / m
    db0 = np.sum(delta2, axis=0) / m
    
    # update weights and biases
    W[2] -= learning_rate * dW2
    b[2] -= learning_rate * db2
    W[1] -= learning_rate * dW1
    b[1] -= learning_rate * db1
    W[0] -= learning_rate * dW0
    b[0] -= learning_rate * db0
    
    return W, b
````

A more detailed implementation:

````python
import numpy as np

# define the sigmoid function
def sigmoid(x, derivative=False):

    if (derivative == True):
        return sigmoid(x,derivative=False) * (1 - sigmoid(x,derivative=False))
    else:
        return 1 / (1 + np.exp(-x))

# choose a random seed for reproducible results
np.random.seed(1)

# learning rate
alpha = .1

# number of nodes in the hidden layer
num_hidden = 3

# inputs
X = np.array([  
    [0, 0, 1],
    [0, 1, 1],
    [1, 0, 0],
    [1, 1, 0],
    [1, 0, 1],
    [1, 1, 1],
])

# outputs
# x.T is the transpose of x, making this a column vector
y = np.array([[0, 1, 0, 1, 1, 0]]).T

# initialize weights randomly with mean 0 and range [-1, 1]
# the +1 in the 1st dimension of the weight matrices is for the bias weight
hidden_weights = 2*np.random.random((X.shape[1] + 1, num_hidden)) - 1
output_weights = 2*np.random.random((num_hidden + 1, y.shape[1])) - 1

# number of iterations of gradient descent
num_iterations = 10000

# for each iteration of gradient descent
for i in range(num_iterations):

    # forward phase
    # np.hstack((np.ones(...), X) adds a fixed input of 1 for the bias weight
    input_layer_outputs = np.hstack((np.ones((X.shape[0], 1)), X))
    hidden_layer_outputs = np.hstack((np.ones((X.shape[0], 1)), sigmoid(np.dot(input_layer_outputs, hidden_weights))))
    output_layer_outputs = np.dot(hidden_layer_outputs, output_weights)

    # backward phase
    # output layer error term
    output_error = output_layer_outputs - y
    # hidden layer error term
    # [:, 1:] removes the bias term from the backpropagation
    hidden_error = hidden_layer_outputs[:, 1:] * (1 - hidden_layer_outputs[:, 1:]) * np.dot(output_error, output_weights.T[:, 1:])

    # partial derivatives
    hidden_pd = input_layer_outputs[:, :, np.newaxis] * hidden_error[: , np.newaxis, :]
    output_pd = hidden_layer_outputs[:, :, np.newaxis] * output_error[:, np.newaxis, :]

    # average for total gradients
    total_hidden_gradient = np.average(hidden_pd, axis=0)
    total_output_gradient = np.average(output_pd, axis=0)

    # update weights
    hidden_weights += - alpha * total_hidden_gradient
    output_weights += - alpha * total_output_gradient

# print the final outputs of the neural network on the inputs X
print("Output After Training: \n{}".format(output_layer_outputs))
````

````python
# Importing the necessary libraries
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
````

````python
"""
Neuron class representing a single neuron
"""

class Neuron:
    def __init__(self, n_inputs, weights=None):
        self.weights = weights if weights else np.random.randn(n_inputs) / (2 * n_inputs)
        self.value = None
        self.gradient = None

    def __call__(self, inputs):
        self.value = np.dot(inputs, self.weights)
        return self.value

    def compute_gradient(self, error):
        self.gradient = error * self.value

    def update_weights(self, learning_rate):
        self.weights -= learning_rate * self.gradient

        # Gradient clipping
        self.gradient = np.clip(self.gradient, -1, 1)
````

````python
"""
NeutralNetwork class representing a neural network
"""

class NeuralNetwork:
    def __init__(self, n_inputs, n_hidden1, n_hidden2, n_output):
        self.hidden1 = [Neuron(n_inputs) for _ in range(n_hidden1)]
        self.hidden2 = [Neuron(n_hidden1) for _ in range(n_hidden2)]
        self.output = Neuron(n_hidden2, weights=None)

    def forward(self, inputs):
        hidden1_outputs = [neuron(inputs) for neuron in self.hidden1]
        hidden2_outputs = [neuron(hidden1_outputs) for neuron in self.hidden2]
        output = self.output(hidden2_outputs)
        return output

    def backward(self, inputs, target, learning_rate):
        output_error = target - self.output.value
        self.output.compute_gradient(output_error)

        for neuron in self.hidden2:
            neuron_error = self.output.gradient * neuron.weights
            neuron.compute_gradient(neuron_error)

        for neuron in self.hidden1:
            neuron_error = np.sum([n.gradient * n.weights[i] for i, n in enumerate(self.hidden2)])
            neuron.compute_gradient(neuron_error)
            for i, input_value in enumerate(inputs):
                neuron.weights[i] -= learning_rate * neuron.gradient * input_value

                # Gradient clipping
                neuron.gradient = np.clip(neuron.gradient, -1, 1)

        for neuron in self.hidden2:
            neuron.update_weights(learning_rate)

        self.output.update_weights(learning_rate)
````

````python
# Set the amount of input data, hidden layers and output data
n_inputs = 3
n_hidden1 = 4
n_hidden2 = 4
n_output = 1

# Creation of a network
network = NeuralNetwork(n_inputs, n_hidden1, n_hidden2, n_output)
````

````python
# Example input data
inputs = np.array([1, -1, 1])
target = 0.5
````

````python
# Training loop
learning_rate = 0.1
for iteration in range(100):
    output = network.forward(inputs)
    network.backward(inputs, target, learning_rate)
    error = target - output
    print(f"Iteration: {iteration+1}, Error: {error}")
````
